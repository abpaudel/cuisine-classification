```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
train = pd.read_json('https://gitlab.com/abpaudel/cuisine/raw/master/cook_train.json')
train.head()
```

```python
all_ingr = []
for i in train.ingredients:
    for word in i:
        all_ingr.append(word)
len(all_ingr)
```

```python
count = Counter(all_ingr)
top10 = dict(count.most_common(15))
top10
```

```python
plt.figure(figsize=(15,7))
plt.bar(range(15), top10.values(), tick_label=list(top10.keys()))
plt.xticks(rotation=45)
```

```python
wc = WordCloud(background_color='white', colormap='Dark2', width=1000, height=500,random_state=2).generate_from_frequencies(count)
plt.figure(figsize=(15,7))

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```
